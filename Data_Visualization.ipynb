```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df=pd.read_csv("/content/merge_data.csv")
dfv = df.copy()
df
```

```python
print("No. of Null cells: ",df.isnull().sum().sum())
```

```python
df.columns
```

Removing Columns that feel not so important:

*   school
*   famsize
*   reason
*   guardian
*   Pstatus








```python
df = df.drop(["Pstatus", "school", 'famsize', 'reason', 'guardian'], axis='columns')
df
```

```python
df['schoolsup']= label_encoder.fit_transform(df['schoolsup'])
df['famsup']= label_encoder.fit_transform(df['famsup'])
df['paid']= label_encoder.fit_transform(df['paid'])
df['activities']= label_encoder.fit_transform(df['activities'])
df['nursery']= label_encoder.fit_transform(df['nursery'])
df['romantic']= label_encoder.fit_transform(df['romantic'])
df['higher']= label_encoder.fit_transform(df['higher'])
df['internet']= label_encoder.fit_transform(df['internet'])
df['Fedu']= label_encoder.fit_transform(df['Fedu'])
df['sex']= label_encoder.fit_transform(df['sex'])
df['address']= label_encoder.fit_transform(df['address'])
df['Mjob']= label_encoder.fit_transform(df['Mjob'])
df['Fjob']= label_encoder.fit_transform(df['Fjob'])
df['address']
```

```python
plt.figure(figsize=(22, 11))
sns.heatmap(df.corr(), annot=True);
```

```python
data = df.copy().drop(['G1','G2'],axis='columns')
corr_train = data.corr()['G3'].abs().sort_values(ascending=False)
sns.barplot(corr_train[1: 20].values, corr_train[1: 20].index)
plt.title("Correlation with Grades.")
```

```python
sns.countplot(df['G3'])
plt.show()
```

```python
data=dfv.copy()
fig = px.histogram(data, x="studytime",
                   width=400, 
                   height=300,
                   histnorm='percent',
                   template="simple_white"
                   )
fig.update_layout(title="studytime", 
                  font_family="San Serif",
                  titlefont={'size': 20},
                  showlegend=True,
                  legend=dict(
                      orientation="v",
                      y=1.0, 
                      yanchor="top", 
                      x=1.0, 
                      xanchor="right"
                  )                 
                 )
fig.update_traces(marker_color=None, marker_line_color='white',
                  marker_line_width=1.5, opacity=0.99)
fig.show()
```

```python
data = dfv.copy()
sex_counts = data["sex"].value_counts()

plt.pie(sex_counts, labels=sex_counts.index,autopct='%1.1f%%')
plt.title("Distribution of Students by Sex")
plt.show()
```

```python
data = dfv.copy()
sns.lmplot(x='G1',y='G2',hue='sex',data=data)
plt.xlabel('G1 ')
plt.ylabel('G2')
plt.title('G1 VS G2')
plt.show()
```

```python
data = dfv.copy()
mjob_counts = data["Mjob"].value_counts()

plt.pie(mjob_counts, labels=mjob_counts.index,autopct='%1.1f%%')
plt.title("Distribution of Students' Mothers' Jobs.")
plt.show()
```

```python
fjob_counts = data["Fjob"].value_counts()

plt.pie(fjob_counts, labels=fjob_counts.index,autopct='%1.1f%%')
plt.title("Distribution of Students' Fathers' Jobs.")
plt.show()
```

```python
data["avg_G12"] = (data["G1"] + data["G2"]) / 2

g=sns.pairplot(data, x_vars=["avg_G12"], y_vars=["G3"], hue="studytime")
g.fig.set_size_inches(8,5)
```
